```python
import pandas as pd

data={
    "Name":["sejel","shetal","ninad"],
"Age":[21,20,22],
"city":["pune","delhi","ratnagiri"]
}
df=pd.DataFrame(data)
print(df)
```

```python
print(df.loc[[0,2]]) # it is labeled based, meaning it uses rows and columns labels
```

```python
print(df.loc[1,"Name"])
```

```python
print(df.loc[1:3,["Name","city"]])
```

```python
print(df.iloc[1]) #it is positional based, meaning it uses integral position of indexing.
```

```python
print(df.iloc[[0,2]])
```

```python
print(df.iloc[0,1])
```

```python
import pandas as pd

data={
    "Name":["sejel","shetal","ninad","sejel","shetal","ninad"],
"Age":[21,20,22,21,20,22,],
"city":["pune","delhi","ratnagiri","pune","delhi","ratanagiri"]
}
df1=pd.DataFrame(data)
print(df1)
```

```python
df_no_duplicates=df1.drop_duplicates()
print(df_no_duplicates)
```

```python
df_no_duplicates=df1.drop_duplicates(subset=["Name"])
print(df_no_duplicates)
```

```python
df_drop_columns=df1.drop(columns=["city"])
df_drop_columns
```

```python
#specific value
df1["Age"]=df1["Age"].replace(21,78)
df1
```

```python
#multiple values
df1["Name"]=df1["Name"].replace({"sejel":"vijay","shetal":"swapnil"})
df1
```

```python
import pandas as pd
import numpy as np
data={
    "Name":["sejel","shetal","ninad","sejel","shetal","ninad"],
"Age":[21,20,np.nan,22,np.nan,34],
"city":["wani","delhi","ratnagiri","pune",None,None]
}
df1=pd.DataFrame(data)
print(df1)
```

```python
df1["city"]=df1["city"].fillna("unknown")
df1
```

```python
pivot=df1.pivot_table(values="Age",index="Name",aggfunc="mean")
pivot
```

```python
grouped=df1.groupby("city").count()
grouped
```
