# 1.Line Graph

```python
x=[2,4,6,8,10]
y=[9,7,5,3,1]


plt.plot(y,x, c ="g", marker = "o")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Line chart of X and Y")
```

```python
x=[2,4,6,8,10]
y=[11,32,15,17,29]
plt.xlabel("X")
plt.ylabel("Y")
plt.title("The graph of Y vs X")
plt.plot(y,x, "b",marker='o',markersize=8)
plt.show()
```

```python
x=[2,4,6,8,10]
y=[1,3,5,7,9]
c= "r"
plt.plot(x,y,c , marker = "d")
plt.show()
```

```python
x=[3,5,9,4,2]
y=[10,6,3,5,7]
c= "g"
plt.plot(y,x,c, marker = "o", linestyle = "--")
plt.show()
```

# 4.Pie chart

```python
# simple data
data = [23,45,100,20,49]
plt.pie(data)
plt.show()
```

```python
data = [23,45,100,20,49]
sub = ["A","B","C","D","E"]
plt.pie(data, labels= sub, autopct= "%.1f%%" )
plt.show()
```

```python
data = [23,45,100,20,49]
sub = ["A","B","c","D","E"]
plt.pie(data, labels= sub ,autopct= "%.1f%%", explode = [0.5,0,0,0.2,0],normalize=True)
plt.show()
```

# 5.Scatter Plot 

```python
import matplotlib.pyplot as plt

# Data
hours_studied = [2, 3, 4, 5, 6, 7, 8]
exam_scores = [100, 600, 750, 900, 950, 700, 600]

# Create a scatter plot
plt.figure(figsize=(15,8))  # Set the size of the figure

plt.scatter(hours_studied, exam_scores, color='lightseagreen', marker='o', label='scatter Points', s = exam_scores )


plt.title('Scatter Plot of Hours Studied vs. Exam Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.legend()  # Display the legend

plt.grid(True)  # Add a grid for better visualization
plt.show()  # Display the plot
```
