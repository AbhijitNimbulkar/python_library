### Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Student': ['Student A', 'Student B', 'Student C', 'Student D'],
    'Test 1 Score': [85, 90, 78, 92],
    'Test 2 Score': [88, 86, 89, 78],
    'Test 3 Score': [92, 80, 85, 88]
}

df = pd.DataFrame(data)
df
```

```python
# Set the 'Student' column as the index
df.set_index('Student', inplace=True)
df
```

```python

# Create a custom heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df, annot = True,cmap = 'magma',fmt = 'd',linewidths=0)  #fmt is formatting of the annotation  d, % , f
#sns.heatmap(df,annot=True,  cmap='YlGnBu', fmt='d', linewidths=0.5)
plt.title('Test Scores Heatmap')
plt.xlabel('Test')
plt.ylabel('Student')
plt.show()
```

```python
sns.heatmap(df, annot=True, cmap='viridis', fmt='d', linewidths=0.2)
```

```python
df= sns.load_dataset('flights')
df
```

```python
df.dtypes
```

```python
flight = df.pivot(index='year', columns='month', values='passengers')
flight
```

```python
plt.figure(figsize=(15,18))
sns.heatmap(flight, annot=True, fmt='d', cmap='viridis_r')
```


- fmt=".0f": Display values as floating-point numbers with zero decimal places.
- fmt=".2f": Display values as floating-point numbers with two decimal places.
- fmt="d": Display values as integers.
- fmt="s": Display values as strings.

## Regression 

- Regplot
- lmplot

```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
# regplot => axes level function
sns.regplot(data = tips, x="total_bill", y = "tip")
plt.show()

#Light bluish shade near to line shows error in line so we can also say that upto that range line may be deviated
```

```python
sns.lmplot(data = tips, x="total_bill", y = "tip")
plt.show()
```
